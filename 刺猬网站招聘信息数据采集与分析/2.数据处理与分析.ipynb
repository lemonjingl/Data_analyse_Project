##### （3） 统计薪资待遇为“面议”的数量，将“面议”的行删除，并将薪资单位转化为“元/天”，求区间的均值。

```python
df1['薪资待遇'].value_counts()
```

```python
# 删除了薪资待遇为面议的
df1=df1[df1['薪资待遇']!='面议']
```

```python
df1['薪资待遇'].value_counts()
```

```python
# 将薪资单位转为“元/天”
def cope(x):
    if '天' in x:
        x=int((int(x.split('元')[0].split('-')[0])+int(x.split('元')[0].split('-')[1]))/2)
        return x
    
    if '月' in x:
        x=int((int(x.split('K')[0].split('-')[0])+int(x.split('K')[0].split('-')[1]))*1000/30)
        return x
    
df1['薪资待遇']=df1['薪资待遇'].apply(cope)
```

##### （4） “发布时间”只保留日期，并提取年份。

```python
# 提取年份
df1['年份']=pd.to_datetime(df1['发布时间']).dt.year
```

##### （5） 将“工作地点”省份名称规范化，如“北京市”、“广西壮族自治区”、“广东省”、“香港特别行政区”等。

```python
# 添加省份列
```

```python
def cope11(x):
    # 删除中间的空格
    x = x.replace(' ', '')
    x=x.replace('线上','')
    x=x.replace('全国','')

    # 根据不同情况进行规范化
    x=x.split('-')[0]

    return x  # 返回处理后的地址
```

```python
# 添加省份列
df1['省份']=df1['公司地址'].apply(cope11)
```

```python
def cope1(x):
    # 删除中间的空格
    x = x.replace(' ', '')
    x = x.replace(' ', '')
    x=x.replace('线上','')
    x=x.replace('全国','')

    # 根据不同情况进行规范化
    if '广西' in x:
        x = x.split('-')[0] + '壮族自治区' + ''.join(x.split('-')[1:])
    elif '新疆' in x:
        x = x.split('-')[0] + '维吾尔自治区' + ''.join(x.split('-')[1:])
    elif '西藏' in x:
        x = x.split('-')[0] + '自治区' + ''.join(x.split('-')[1:])
    elif '宁夏' in x:
        x = x.split('-')[0] + '回族自治区' + ''.join(x.split('-')[1:])
    elif '香港' in x or '澳门' in x:
        x = x.split('-')[0] + '特别行政区' + ''.join(x.split('-')[1:])
    else:
        x = x.split('-')[0] + '省' + ''.join(x.split('-')[1:])

    return x  # 返回处理后的地址


df1['公司地址'] = df1['公司地址'].apply(cope1)
```

#### 3.分词和去停用词

##### 对“工作描述”进行分词处理，去除停用词，并保存结果数据

```python
import jieba
```

```python
# 定义停用词
stop_words=set(['的', '和', '是', '在', '与', '这', '有', '为', '进行', '对','工作','职责','岗位','要求','等','将','获得','你'])

# 定义分词和去停用词的函数
def tokenize_and_filter(text):
    words=jieba.cut(text)# 分词
    filtered_words=[word for word in words if word not in stop_words]
    return ' '.join(filtered_words)
```

```python
df1['工作描述']=df1['工作描述'].apply(tokenize_and_filter)
```
